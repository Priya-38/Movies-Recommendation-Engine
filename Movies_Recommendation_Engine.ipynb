Importing the libraries:

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Preprocessing:

```python
#import csv(comma separator values) file into pandas dataframe:
movies = pd.read_csv('/content/drive/MyDrive/movies.csv')
```

```python
#first five rows:
movies.head()
```

```python
#last five rows:
movies.tail()
```

```python
#number of rows and columns:
movies.shape
```

```python
#checking the missing values:
movies.isnull().sum()
```

```python
#selecting the relevant features in our dataset:
selected_features = ['genres','cast','tagline','keywords','director']
print(selected_features)
```

```python
#replacing all the missing values in our relevant columns by null string:
for feature in selected_features:
  movies[feature] = movies[feature].fillna(' ')
```

```python
#again checking the missing values:
movies.isnull().sum()
```

```python
#combining the selected features:
combined_features = movies['genres'] + movies['cast'] + movies['keywords'] + movies['director'] + movies['tagline']
print(combined_features)
```

```python
#converting the text data into feature vectors:
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
#getting the similarity score using the cosine similarity:
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
#getting the movie name from the user:
movie_name = input("Enter movie name")
```

```python
#creating a list of all the movies in the dataset:
list_of_all_titles = movies['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match of the movie name given by the user:
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = movies[movies.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting the list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on theie similarity score
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#printing the name of the similar movies based on the index

print('Movies Suggested for you: \n')

i = 1
for movie in sorted_similar_movies:
   index = movie[0]
   title_from_index = movies[movies.index==index]['title'].values[0]
   if (i<21):
     print (i, title_from_index)
     i = i+1
```

```python
#making the recommendation system

#getting the movie name from the user

movie_name = input('Enter the movie name: ')

#creating a list of all the movies in the dataset

list_of_all_titles = movies['title'].tolist()

#finding the close match of the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies[movies.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

#sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)


#printing the name of the similar movies based on the index

print('Movies Suggested for you: \n')

i = 1
for movie in sorted_similar_movies:
   index = movie[0]
   title_from_index = movies[movies.index==index]['title'].values[0]
   if (i<21):
     print (i,':', title_from_index)
     i = i+1
```
